# Assignment 1: Working with NumPy

Objective: Understand NumPy arrays and operations.

```
import numpy as np
```

## 1. 1D Array Operations

```

arr = np.arange(1, 21)
arr, arr.sum(), arr.mean(), np.median(arr), arr.std()
```

```
np.where(arr > 10)
```

## 2. 2D Array Operations

```

arr2d = np.arange(1, 17).reshape(4, 4)
arr2d
```

```

arr2d.T, arr2d.sum(axis=1), arr2d.sum(axis=0)
```

## 3. Random Arrays

```

a = np.random.randint(1, 21, (3,3))
b = np.random.randint(1, 21, (3,3))
a, b, a+b, a-b, a*b, np.dot(a,b)
```

## 4. Reshape and Slice

```

arr1d = np.arange(1,13)
reshaped = arr1d.reshape(3,4)
reshaped[:2, -2:]
```
